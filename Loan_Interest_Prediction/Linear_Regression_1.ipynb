```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# Predictions on the training set
pred_train = linear_regression.predict(x_train)

# Predictions on the test set
pred_test = linear_regression.predict(x_test)

# Function to calculate MAPE
def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# Calculate metrics for the training set
rmse_train = np.sqrt(mean_squared_error(y_train, pred_train))
mae_train = mean_absolute_error(y_train, pred_train)
r2_train = r2_score(y_train, pred_train)

# Calculate adjusted R-squared for the training set
n_train = len(y_train)
k_train = x_train.shape[1]
adj_r2_train = 1 - ((1 - r2_train) * (n_train - 1) / (n_train - k_train - 1))

# Calculate MAPE for the training set
mape_train = mean_absolute_percentage_error(y_train, pred_train)

# Print metrics for the training set
print("Training Set Metrics:")
print("RMSE:", rmse_train)
print("MAE:", mae_train)
print("R-squared:", r2_train)
print("Adjusted R-squared:", adj_r2_train)
print("MAPE:", mape_train)

# Calculate metrics for the test set
rmse_test = np.sqrt(mean_squared_error(y_test, pred_test))
mae_test = mean_absolute_error(y_test, pred_test)
r2_test = r2_score(y_test, pred_test)

# Calculate adjusted R-squared for the test set
n_test = len(y_test)
k_test = x_test.shape[1]
adj_r2_test = 1 - ((1 - r2_test) * (n_test - 1) / (n_test - k_test - 1))

# Calculate MAPE for the test set
mape_test = mean_absolute_percentage_error(y_test, pred_test)

# Print metrics for the test set
print("\nTest Set Metrics:")
print("RMSE:", rmse_test)
print("MAE:", mae_test)
print("R-squared:", r2_test)
print("Adjusted R-squared:", adj_r2_test)
print("MAPE:", mape_test)
```
